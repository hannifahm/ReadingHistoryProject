```python
import os
import pandas as pd
```

```python
pwd = os.getcwd()

filepath = pwd + '\\BookData.csv'
```

```python
df = pd.read_csv(filepath)
```

```python
df['Genre'] = df['Genre'].str.split(", ")
df['Moods'] = df['Moods'].str.split(", ")
```

```python
df = df.explode('Genre')
df
```

```python
df = df.explode('Moods')
df
```

```python
df.to_csv(pwd + '\\FinalBookData.csv', index=False)
```
